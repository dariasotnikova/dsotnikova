```python
d = open('капитанская дочка.txt', 'r', encoding='utf-8')
```

```python
text = d.read()
```

```python
words = [word.strip("!?.,:;«»-") for word in text.split()]
```

```python
text2= [w for w in text.split() if w.isalnum()]
print(text2)
```

```python
from collections import Counter
counts = Counter(words) 
print(counts)
```

```python
text3=' '.join(text2)
print(text3)
```

```python
from pymystem3 import Mystem 
text = text3
m = Mystem()
lemmas = m.lemmatize(text)

print(type(lemmas))
print(''.join(lemmas))
```

```python
clean_lemmas = [w for w in lemmas if w.isalpha()]
```

```python
list = {}
for word in text.split():
    word= word.lower()
    if word in list:
        value = list[word]
        list[word]=value+1
    else:
        list[word]=1
    
print(list) 
```

```python
f = open('rus_stopwords.txt', 'r', encoding='utf-8')
```

```python
stop = f.read().split('\n')
```

```python
no_stop_words = [w for w in clean_lemmas if w not in stop]
new_counts = Counter(no_stop_words)
print(new_counts.most_common(100))
```

```python
bigrams = []
for i in range (len(clean_lemmas)-1):
    bigram = (clean_lemmas [i], clean_lemmas[i+1])
    bigrams.append(bigram)
    
bigram_counts = Counter(bigrams)
print(bigram_counts.most_common(100))
```
