```python
from pymystem3 import Mystem
m = Mystem()
with open('Chechov.txt', 'r', encoding='utf-8') as d:
    text= d.read()
```

```python
import json
story = m.analyze(text)
with open('story.json','w', encoding='utf-8') as f:
    json.dump(story, f, ensure_ascii=False)
```

```python
with open('story.json', 'r', encoding='utf-8') as f:
    story = json.loads(f.read())
```

```python
words = {}
for word in story:
    try:
        form = word['text']
        gr = word['analysis'][0]['gr']
        words[form]=gr
    except (KeyError, IndexError) as e:
        pass
print(words)
```

```python
import re
pos = re.compile('[A-Z]+')
words_pos = {}
for k, v in words.items():
        words_pos[k] = pos.match(v).group(0)
        print(words_pos)
```
