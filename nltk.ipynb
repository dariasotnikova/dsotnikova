```python
f = open ('Alice.txt', 'r', encoding='utf-8') 
text = f.read()
punctuation= [w for w in text.split() if w.isalnum()]
new_text=' '.join(punctuation)
down_register= [w for w in new_text.split() if w.islower()]
text2= ' '.join(down_register)
from nltk.tokenize import word_tokenize
[b for b in nltk.bigrams(word_tokenize(text))]
nltk.trigrams(word_tokenize(text))
from pymystem3 import Mystem 
m = Mystem()
lemmas = m.lemmatize(text2)
type(lemmas) 
text3= ' '.join(lemmas)
d = open('rus_stopwords.txt', 'r', encoding='utf-8') 
stop_words = d.read().split('\n') 
without_stop_words = [w for w in text3.split() if w not in stop_words] 
clean_text = ' '.join(without_stop_words) 
from collections import Counter
new_counts = Counter(clean_text)
print(new_counts.most_common(30)) 
```

```python
# X и Y - координаты точек
X = [1,   1.5, 2,   3, 5, 6]
Y = [1.8, 1,   3.6, 2, 1, 3]

plt.plot(X, Y) # рисуем график - последовательно соединяем точки с координатами из X и Y
plt.show()
```

```python
plt.scatter(X, Y, s=[20, 60, 30, 100, 60, 10], c='plum', marker='^')
plt.show()
```

```python
markers = ['o', '*', '8', 'D', '^', 'P']

for x, y, m in zip(X, Y, markers):
    plt.scatter([x],[y], marker=m, s=100)
plt.show()
```

```python
dots = ['Первая', "Вторая", "Третья", "Четвертая", "Пятая", "Шестая"]

for x, y, m, d in zip(X, Y, markers, dots):
    plt.scatter(x, y, marker=m, s=100)
    plt.text(x+0.1, y+0.1, d) # +0.1 - это чтобы текст не наползал на маркер, а отрисовывался чуть выше и правее 
plt.show()
```

```python
f = open ('Alice.txt', 'r', encoding='utf-8') 
text = f.read()
words = len(text)
print(words)
punctuation= [w for w in text.split() if w.isalnum()]
new_text=' '.join(punctuation)
```

```python
from pymystem3 import Mystem 
m = Mystem()
lemmas = m.lemmatize(new_text)
type(lemmas) 
```

```python
import nltk
from nltk.corpus import stopwords
sw = stopwords.words('english')
filtered = [w.lower() for w in lemmas if w.lower() not in sw and w != ' ']
print(filtered)
print(len(filtered))
```

```python
from collections import Counter
new_counts = Counter(filtered)
print(new_counts.most_common(30))
```
