```python
penguin = [ 
  '   _~_    ',
  '  (o o)   ',
  ' /  V  \\  ',
  '/(  _  )\\ ',
  '  ^^ ^^   ' ] 

quantity = int(input("Введите любое число от 1 до 9\n"))
for i in penguin: 
    print(quantity * i, sep = '')
```
