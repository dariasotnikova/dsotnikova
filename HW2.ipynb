# Задание 1
Дан текст:

>варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг

**Необходимо выделить все биграммы в этом тексте и сохранить их список в какую-либо переменную. Биграммы -- это пары стоящих рядом слов, например, "варкалось хливкие", "хливкие шорьки" и т.д.**

**Подсказка №1**: текст нужно разбить на слова с помощью функции split(), а потом использовать цикл for для создания списка биграмм. На каждой итерации цикла нужно печатать биграмму, а потом добавлять ее в список.

**Подсказка №2**: количество биграмм в тексте = количество слов в тексте - 1.

```python
sentence = """варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг"""
words = sentence.split() #разбиваем предложение на слова
bigrams = [] #создаем пустой список
for i in range(len(words)-1):
    bigrams.append([words[i],words[i+1]]) #добавляем биграммы в список, где i - некоторое заданное число
print(bigrams)
```

```python
#еще попробовала сделать через list comprehension 

bigrams=[[sentence.split()[i], sentence.split()[i+1]] for i in range(len(sentence.split())-1)]
print(bigrams)
```

```python
poem= """Удивительная кошка

Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!""" #создаем переменную 

print(len(poem)) #количество символов в стихотворении 

words = [word.strip("-,.!") for word in poem.split()] #обрезаем пунктуацию с помощью list comprehensions
print(words)

poem2=' '.join(words) #объединяем получившиеся слова через пробел, используя функцию join
print(poem2)

print([w.lower() if w.istitle() else w for w in poem2.split()]) #переводим символы верхнего регистра в нижний 

print(len(words)) #длина текста в словах

uniquewords=set(poem.split()) #создаем множество, чтобы узнать количество уникальных элементов
print(len(uniquewords))

print(words[-10:]) #последние 10 слов с помощью среза

string='\n'.join(words) #объединим слова в строку и перенесем каждое построчно
print(string)

print(string[19:29:]) #20 и 30 символы строки с помощью среза

bigrams=[[poem.split()[i], poem.split()[i+1]] for i in range(len(poem.split())-1)] 
print(bigrams)

print(len(bigrams)) #посчитаем количество биграмм

quantity = [] #создаем пустой список
for i in range(len(bigrams)):
    quantity.append(bigrams.count(bigrams[i])) #счиатем количесвто биграмм с помощью метода count()
print(quantity)







```
